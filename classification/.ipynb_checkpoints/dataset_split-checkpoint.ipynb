```python
#created by Cheng Zida
#2019.6.26
'''
This program is used to select useful imgs,remove redundancy and split them from /DATA5_DB8/data/zdcheng/hyperthyreosis_eye
这个程序用于从 /DATA5_DB8/data/zdcheng/hyperthyreosis_eye的三批数据中选出能用的图片（已在selected文件夹中），
去重并将同一个人的图片归为一个文件夹（split_id文件夹中），最后分为训练、验证、测试集
'''
```

```python
import os
import json
import random
import copy
```

```python
dataroot = '/Users/zdcheng/数据集/hyperthyreosis_eye/selected'
split_id_root = '/Users/zdcheng/数据集/hyperthyreosis_eye/split_id'
```

```python
#处理 split_id目录，得到两个list，分别表示单个图片的list和多个图片来自同一个人的list
singleimg_list = []
multiimg_list = []
for root,dirs,files in os.walk(split_id_root):
    if root == split_id_root:
        singleimg_list += files
    else:
        tmp = [i for i in files if i.endswith('jpg') or i.endswith('JPG')]
        multiimg_list.append(tuple(tmp))
    
```

```python
singleimg_list = [i for i in singleimg_list if i.endswith('jpg') or i.endswith('JPG')]
```

```python
singleimg_list.sort()
```

```python
len(singleimg_list)
```

```python
len(multiimg_list)
```

```python
#共约151个不同的人，选择其中的31个人作为测试集，其中16个多图片的人，15个单图片的人
#剩余120个人，60多图片，60单图片，按15个一组分为4组
trainval_list = []
test_list = []

random.seed(15)
random.shuffle(multiimg_list)
random.seed(15)
random.shuffle(singleimg_list)
person_id = 0

# 处理多图片的人
for idx,i in enumerate(multiimg_list):
    #前16个用于测试集
    if idx <= 15:
        for img in i:
            tmp_dict = dict()
            tmp_dict['img'] = img
            tmp_dict['person_id'] = person_id
            test_list.append(tmp_dict)
        person_id += 1
    #train_val集
    else :
        for img in i:
            tmp_dict = dict()
            tmp_dict['img'] = img
            tmp_dict['person_id'] = person_id
            #idx 16~30 fold0
            if idx <= 30:
                tmp_dict['fold'] = 0 
            elif idx <= 45:
                tmp_dict['fold'] = 1
            elif idx <= 60:
                tmp_dict['fold'] = 2
            else:
                tmp_dict['fold'] = 3
            trainval_list.append(tmp_dict)
        person_id += 1
# 处理单图片的人
for idx,i in enumerate(singleimg_list):
    #前15个用于测试集
    if idx <= 14:
        img = i
        tmp_dict = dict()
        tmp_dict['img'] = img
        tmp_dict['person_id'] = person_id
        person_id += 1
        test_list.append(tmp_dict)
    #train_val集
    else :
        img = i
        tmp_dict = dict()
        tmp_dict['img'] = img
        tmp_dict['person_id'] = person_id
        person_id += 1
        #idx 16~30 fold0
        if idx <= 29:
            tmp_dict['fold'] = 0 
        elif idx <= 44:
            tmp_dict['fold'] = 1
        elif idx <= 59:
            tmp_dict['fold'] = 2
        else:
            tmp_dict['fold'] = 3
        trainval_list.append(tmp_dict)
```

```python
with open('trainval_pos.json','w') as f:
    json.dump(trainval_list,f,indent=4)
```

```python
with open('test_pos.json','w') as f:
    json.dump(test_list,f,indent=4)
```

```python
'''
下面处理正常人图像，在亚洲人中每个性别各选1500张
'''
normaldata_root = '/Users/zdcheng/数据集/hyperthyreosis_eye/SCUT-FBP5500_v2/Images'
normal_all_lis = os.listdir(normaldata_root)
```

```python
AM_lis = [i for i in normal_all_lis if i.startswith('AM')]
AF_lis = [i for i in normal_all_lis if i.startswith('AF')]
```

```python
random.seed(15)
random.shuffle(AM_lis)
random.seed(15)
random.shuffle(AF_lis)
```

```python
sel_AM_lis = AM_lis[:1500]
sel_AF_lis = AF_lis[:1500]
sel_normal_lis = sel_AM_lis + sel_AF_lis #包含3000个图片
```

```python
random.seed(30)
random.shuffle(sel_normal_lis)

normal_trainval_list = []
normal_test_list = []
for idx,i in enumerate(sel_normal_lis):
    tmp_dict = dict()
    tmp_dict['img'] = i
    #600 test
    if idx < 600:
        normal_test_list.append(tmp_dict)
    elif idx < 1200:
        tmp_dict['fold'] = 0
        normal_trainval_list.append(tmp_dict)
    elif idx < 1800:
        tmp_dict['fold'] = 1
        normal_trainval_list.append(tmp_dict)
    elif idx < 2400:
        tmp_dict['fold'] = 2
        normal_trainval_list.append(tmp_dict)
    else:
        tmp_dict['fold'] = 3
        normal_trainval_list.append(tmp_dict)
```

```python
all_trainval_list = []
all_test_list = []
for i in trainval_list:
    new_dict = copy.deepcopy(i)
    new_dict['label'] = 1
    all_trainval_list.append(new_dict)
for i in test_list:
    new_dict = copy.deepcopy(i)
    new_dict['label'] = 1
    all_test_list.append(new_dict)
for i in normal_trainval_list:
    new_dict = copy.deepcopy(i)
    new_dict['label'] = 0
    all_trainval_list.append(new_dict)
for i in normal_test_list:
    new_dict = copy.deepcopy(i)
    new_dict['label'] = 0
    all_test_list.append(new_dict)
random.seed(40)
random.shuffle(all_trainval_list)
random.seed(45)
random.shuffle(all_test_list)
```

```python
with open('all_trainval.json','w') as f:
    json.dump(all_trainval_list,f,indent=4)
with open('all_test.json','w') as f:
    json.dump(all_test_list,f,indent=4)
```
