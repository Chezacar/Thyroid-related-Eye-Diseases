```python
### convert 'pts' to 'txt' format
import os
import re
import struct
import shutil
import cv2
```

```python
def get_files(dname, suffix):
    pts_list = []
    for fname in os.listdir(dname):
        if fname.endswith(suffix):
            pts_list += [fname]
    return pts_list


def pts2txt(din, dout, src):
    src_p = os.path.join(din, src)
    data = open(src_p, 'rb').read()
    if len(data) < 692:
        print('Warning', src)
        return 0
    points = struct.unpack('i172f', data)
    # print points

    dst = src
    dst = dst.replace('pts', 'txt')
    dst_p = os.path.join(dout, dst)
    # print dst_p

    fout = open(dst_p, 'w')
    pnum = len(points[1:])
    for i in range(1, pnum, 2):
        fout.write('%f ' % points[i])
        fout.write('%f\n' % points[i + 1])
    fout.close()

    return 1
```

```python
src = '/DATA5_DB8/data/zdcheng/hyperthyreosis_eye/SCUT-FBP5500_v2/facial landmark'
dst = '/DATA5_DB8/data/zdcheng/hyperthyreosis_eye/SCUT-FBP5500_v2/landmark_txt'

if not os.path.exists(dst):
    os.mkdir(dst)

pts_list = get_files(src, 'pts')
```
