```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import copy
```

```python
point_txt = '/DATA5_DB8/data/zdcheng/hyperthyreosis_eye/SCUT-FBP5500_v2/landmark_txt/AF1.txt'
img = '/DATA5_DB8/data/zdcheng/hyperthyreosis_eye/SCUT-FBP5500_v2/Images/AF1.jpg'
```

```python
point_txt = open(point_txt,'r')
lines = point_txt.readlines()
points = []
for i in range(len(lines)):
    x = int(float(lines[i].split(' ')[0]))
    y = int(float(lines[i].split(' ')[1]))
    points.append((x,y))
img = cv2.imread(img)
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize = (20,20))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
img_RGBpt = copy.deepcopy(img_RGB)

white = (255,255,255)
for idx,p in enumerate(points):
    if idx > 59:
        break
    cv2.circle(img_RGBpt,p,2,white,-1)
    
ax1.imshow(img_RGB)
ax2.imshow(img_RGBpt)
```
