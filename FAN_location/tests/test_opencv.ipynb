```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import face_alignment
```

```python
def plt_show(img):
    tmp = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(tmp)
def pad(img):
    h,w,_ = img.shape
#     h,w = (h//2,w//2)
    h,w = (h,w)
    return cv2.copyMakeBorder(img,h,h,w,w,cv2.BORDER_CONSTANT)
```

```python
img_path = '/DATA5_DB8/data/zdcheng/hyperthyreosis_eye/eyelid/1.JPG'
```

```python
img = cv2.imread(img_path)
img_padded = pad(img)
img_padded_RGB = cv2.cvtColor(img_padded, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_padded_RGB)
```

```python
ratio = 3
tmp_h,tmp_w,_ = img_padded_RGB.shape
input = cv2.resize(img_padded_RGB,(tmp_w//ratio,tmp_h//ratio))
```

```python
plt.imshow(input)
```

```python
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False)
preds = fa.get_landmarks(input)
```

```python
pred = preds[0]
```
